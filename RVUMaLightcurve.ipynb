```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
r = pd.read_csv('RVUMa_R.csv')
g = pd.read_csv('RVUMa_G.csv')
b = pd.read_csv('RVUMa_B.csv')
```

```python
fig = plt.figure(figsize = (15,7))

fig = plt.figure(figsize=(15,7))

# plot red data series
plt.errorbar(r['phaseII'], r['mag'], yerr=r['error'], fmt='x', color='black', label = 'R')

# plot green data series
plt.errorbar(g['phaseII'], g['mag'], yerr=g['error'], fmt='+', color='grey', label = 'G')

# plot blue data series
plt.errorbar(b['phaseII'], b['mag'], yerr=b['error'], fmt='*', color='dimgrey', label = 'B')

plt.gca().invert_yaxis()
plt.legend()
plt.xlabel("$\phi$")
plt.ylabel("Instrumental Magnitude")
```
