```python
pd.options.display.float_format = '{:.10f}'.format

def JD(df, column_name):
    # function to convert our datetime values to Julian date
    # inputs the data file name and the column name 
    
    def datetimeJD(dt): 
        # Calculate the Julian date using the formula:
        # JD = 2451545.0 + (UNIX time in seconds - 946684800) / 86400
        # UNIX time is the time in second since Jan 1st 1970 00:00:00 UTC time 
        # dt is used to return numpy array of python datetime.
        
        unix_time = dt.timestamp() #returns time from the 1970 date
        jd = 2451545.0 + (unix_time - 946684800) / 86400
        
        return jd
    
    # declares a list for the dates 
    julian_dates = []
    
    # for loop to go through the entire column and convert each date and time:
    for date_str in df[column_name]:
        # the strptime function creates a datetime object from the given string
        dt = datetime.strptime(date_str, '%Y-%m-%d %H:%M:%S') 
        jd = datetimeJD(dt)
        julian_dates.append(jd)

    date_df['julian_date'] = julian_dates
    # creates column 'julian date' to store the values in

    return date_df

# Converting our 'time' column to Julian date in our original data frame
date_df = JD(date_df, 'time')


# Print the new dataframe with the Julian dates
date_df
```
